```python
%pip install xgboost
%pip install lightgbm
%pip install catboost

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, GroupKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as imbPipeline
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
X_train = pd.read_csv('/content/asd_multimodal_train_features.csv')
X_test = pd.read_csv('/content/asd_multimodal_test_features.csv')
y_train = pd.read_csv('/content/asd_train_labels.csv').values.ravel()
y_test = pd.read_csv('/content/asd_test_labels.csv').values.ravel()

# Use participant IDs for group-based validation
if 'Participant_ID' in X_train.columns and 'Participant_ID' in X_test.columns:
    train_participants = X_train['Participant_ID'].values
    test_participants = X_test['Participant_ID'].values

    # Drop 'Participant_ID' from features for training
    X_train = X_train.drop(columns=['Participant_ID'])
    X_test = X_test.drop(columns=['Participant_ID'])
else:
    # If Participant_ID does not exist, use index as group
    train_participants = X_train.index.values
    test_participants = X_test.index.values

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Handle class imbalance using SMOTE
smote = SMOTE(random_state=42, k_neighbors=5)
X_train_res, y_train_res = smote.fit_resample(X_train_scaled, y_train)

# Initialize models with class weighting for imbalance
models = {
    "Random Forest": RandomForestClassifier(n_estimators=200,
                                           class_weight='balanced_subsample',
                                           random_state=42),
    "XGBoost": XGBClassifier(scale_pos_weight=np.sum(y_train == 0)/np.sum(y_train == 1),
                             random_state=42),
    "LightGBM": LGBMClassifier(class_weight='balanced', random_state=42),
    "Logistic Regression": LogisticRegression(class_weight='balanced',
                                             max_iter=1000,
                                             solver='saga',
                                             random_state=42),
    "SVM": SVC(class_weight='balanced',
              probability=True,
              random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=150,
                                                   random_state=42),
    "Neural Network": MLPClassifier(hidden_layer_sizes=(100, 50),
                                 early_stopping=True,
                                 random_state=42),
    "AdaBoost": AdaBoostClassifier(n_estimators=100, random_state=42),
    "CatBoost": CatBoostClassifier(silent=True, auto_class_weights='Balanced')
}

# Evaluation function
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    metrics = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'F1-Score': f1_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'ROC-AUC': roc_auc_score(y_test, y_proba)
    }

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Non-ASD', 'ASD'],
                yticklabels=['Non-ASD', 'ASD'])
    plt.title(f'{type(model).__name__} Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    # Classification report
    print(classification_report(y_test, y_pred, target_names=['Non-ASD', 'ASD']))

    return metrics

# Train and evaluate models
results = {}
best_model = None
best_f1 = 0

for name, model in models.items():
    print(f"\n{'='*50}")
    print(f"Training and evaluating {name}")
    print(f"{'='*50}")

    # Train model
    model.fit(X_train_res, y_train_res)

    # Evaluate
    metrics = evaluate_model(model, X_test_scaled, y_test)
    results[name] = metrics

    # Check if best model
    if metrics['F1-Score'] > best_f1:
        best_f1 = metrics['F1-Score']
        best_model = model

    print(f"{name} Performance:")
    for metric, value in metrics.items():
        print(f"{metric}: {value:.4f}")

# Hyperparameter tuning for best model
if best_model and isinstance(best_model, RandomForestClassifier):
    print("\nTuning Random Forest hyperparameters...")
    param_grid = {
        'n_estimators': [100, 200, 300],
        'max_depth': [10, 20, 30, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }

    rf_tuned = GridSearchCV(
        RandomForestClassifier(class_weight='balanced_subsample', random_state=42),
        param_grid,
        cv=StratifiedKFold(n_splits=5),
        scoring='f1',
        n_jobs=-1
    )
    rf_tuned.fit(X_train_res, y_train_res)

    print(f"Best parameters: {rf_tuned.best_params_}")
    print(f"Best F1-Score: {rf_tuned.best_score_:.4f}")

    # Evaluate tuned model
    print("Tuned Random Forest Performance:")
    tuned_metrics = evaluate_model(rf_tuned, X_test_scaled, y_test)
    results['Tuned Random Forest'] = tuned_metrics

# Group-based cross-validation
print("\nGroup-based Cross-Validation with LightGBM")
group_kfold = GroupKFold(n_splits=5)
lgb_model = LGBMClassifier(class_weight='balanced', random_state=42)

group_scores = {'Accuracy': [], 'F1-Score': [], 'ROC-AUC': []}

for train_idx, val_idx in group_kfold.split(X_train, y_train, groups=train_participants):
    X_train_fold, X_val_fold = X_train_scaled[train_idx], X_train_scaled[val_idx]
    y_train_fold, y_val_fold = y_train[train_idx], y_train[val_idx]

    # Handle class imbalance for this fold
    X_train_res_fold, y_train_res_fold = smote.fit_resample(X_train_fold, y_train_fold)

    # Train and evaluate
    lgb_model.fit(X_train_res_fold, y_train_res_fold)
    y_pred = lgb_model.predict(X_val_fold)
    y_proba = lgb_model.predict_proba(X_val_fold)[:, 1]

    group_scores['Accuracy'].append(accuracy_score(y_val_fold, y_pred))
    group_scores['F1-Score'].append(f1_score(y_val_fold, y_pred))
    group_scores['ROC-AUC'].append(roc_auc_score(y_val_fold, y_proba))

print("\nGroup-based Cross-Validation Results:")
for metric, scores in group_scores.items():
    print(f"Mean {metric}: {np.mean(scores):.4f} ± {np.std(scores):.4f}")

# Feature importance visualization
if best_model:
    try:
        if hasattr(best_model, 'feature_importances_'):
            importances = best_model.feature_importances_
            feature_names = X_train.columns

            # Create feature importance dataframe
            fi_df = pd.DataFrame({
                'Feature': feature_names,
                'Importance': importances
            }).sort_values('Importance', ascending=False)

            # Plot top 15 features
            plt.figure(figsize=(12, 8))
            sns.barplot(x='Importance', y='Feature', data=fi_df.head(15))
            plt.title('Top 15 Important Features')
            plt.tight_layout()
            plt.show()

            # Save feature importances
            fi_df.to_csv('feature_importances.csv', index=False)
    except:
        print("Could not extract feature importances")

# Compare model performances
results_df = pd.DataFrame(results).T
results_df = results_df.sort_values('F1-Score', ascending=False)

print("\nModel Performance Comparison:")
print(results_df)

# Plot model comparison
plt.figure(figsize=(12, 8))
results_df.plot(kind='bar', y=['Accuracy', 'F1-Score', 'ROC-AUC'])
plt.title('Model Performance Comparison')
plt.ylabel('Score')
plt.xticks(rotation=15)
plt.legend(loc='lower right')
plt.tight_layout()
plt.savefig('model_comparison.png')
plt.show()

# Save best model
import joblib
joblib.dump(best_model, 'best_asd_model.pkl')
print(f"Saved best model with F1-Score: {best_f1:.4f}")

# Generate predictions for analysis
predictions = best_model.predict(X_test_scaled)
probas = best_model.predict_proba(X_test_scaled)[:, 1]

# Create results dataframe
results_df = pd.DataFrame({
    'Participant_ID': test_participants,
    'True_Label': y_test,
    'Predicted_Label': predictions,
    'ASD_Probability': probas
})

# Save detailed results
results_df.to_csv('asd_predictions_with_probabilities.csv', index=False)
print("Saved detailed predictions")
```

```python
# Install required libraries
%pip install xgboost lightgbm catboost imbalanced-learn scikit-learn pandas numpy matplotlib seaborn

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, GroupKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as imbPipeline
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets (update paths as needed)
X_train = pd.read_csv('/content/asd_multimodal_train_features.csv')
X_test = pd.read_csv('/content/asd_multimodal_test_features.csv')
y_train = pd.read_csv('/content/asd_train_labels.csv').values.ravel()
y_test = pd.read_csv('/content/asd_test_labels.csv').values.ravel()
```

```python
# Handle participant IDs
train_participants = X_train['Participant_ID'] if 'Participant_ID' in X_train.columns else X_train.index.values
test_participants = X_test['Participant_ID'] if 'Participant_ID' in X_test.columns else X_test.index.values

X_train = X_train.drop(columns=['Participant_ID'], errors='ignore')
X_test = X_test.drop(columns=['Participant_ID'], errors='ignore')

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
smote = SMOTE(random_state=42, k_neighbors=5)
X_train_res, y_train_res = smote.fit_resample(X_train_scaled, y_train)
```

```python
# Calculate class weight for XGBoost
scale_pos_weight = np.sum(y_train == 0) / np.sum(y_train == 1) if np.sum(y_train == 1) > 0 else 1

models = {
    "Random Forest": RandomForestClassifier(n_estimators=200, class_weight='balanced_subsample', random_state=42),
    "XGBoost": XGBClassifier(scale_pos_weight=scale_pos_weight, random_state=42, eval_metric='logloss'),
    "LightGBM": LGBMClassifier(class_weight='balanced', random_state=42),
    "Logistic Regression": LogisticRegression(class_weight='balanced', max_iter=1000, solver='saga', random_state=42),
    "SVM": SVC(class_weight='balanced', probability=True, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=150, random_state=42),
    "Neural Network": MLPClassifier(hidden_layer_sizes=(100, 50), early_stopping=True, random_state=42),
    "AdaBoost": AdaBoostClassifier(n_estimators=100, random_state=42),
    "CatBoost": CatBoostClassifier(silent=True, auto_class_weights='Balanced', random_state=42)
}
```

```python
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else [0]*len(y_test)

    metrics = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'F1-Score': f1_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'ROC-AUC': roc_auc_score(y_test, y_proba) if len(np.unique(y_test)) > 1 else 0.5
    }

    # Confusion Matrix
    plt.figure(figsize=(6,4))
    sns.heatmap(confusion_matrix(y_test, y_pred),
                annot=True, fmt='d', cmap='Blues',
                xticklabels=['Non-ASD', 'ASD'],
                yticklabels=['Non-ASD', 'ASD'])
    plt.title(f'{type(model).__name__} Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    print(classification_report(y_test, y_pred, target_names=['Non-ASD', 'ASD']))
    return metrics
```

```python
results = {}
best_model = None
best_f1 = 0

for name, model in models.items():
    print(f"\n{'='*50}")
    print(f"Training {name}")
    print(f"{'='*50}")

    try:
        model.fit(X_train_res, y_train_res)
        metrics = evaluate_model(model, X_test_scaled, y_test)
        results[name] = metrics

        if metrics['F1-Score'] > best_f1:
            best_f1 = metrics['F1-Score']
            best_model = model
            best_model_name = name

        print(f"{name} Performance:")
        for metric, value in metrics.items():
            print(f"{metric}: {value:.4f}")

    except Exception as e:
        print(f"Error with {name}: {str(e)}")
        results[name] = None
```

```python
if best_model and 'Random Forest' in best_model_name:
    print("\nTuning Random Forest...")
    param_grid = {
        'n_estimators': [100, 200, 300],
        'max_depth': [10, 20, 30, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }

    grid_search = GridSearchCV(
        RandomForestClassifier(class_weight='balanced_subsample', random_state=42),
        param_grid,
        cv=StratifiedKFold(n_splits=5),
        scoring='f1',
        n_jobs=-1
    )
    grid_search.fit(X_train_res, y_train_res)
    best_model = grid_search.best_estimator_
    print(f"Best params: {grid_search.best_params_}")
```

```python
print("\nGroup-based CV with LightGBM:")
group_kfold = GroupKFold(n_splits=5)
lgbm = LGBMClassifier(class_weight='balanced', random_state=42)
metrics_cv = {'Accuracy': [], 'F1-Score': [], 'ROC-AUC': []}

for train_idx, val_idx in group_kfold.split(X_train, y_train, groups=train_participants):
    X_tr, X_val = X_train_scaled[train_idx], X_train_scaled[val_idx]
    y_tr, y_val = y_train[train_idx], y_train[val_idx]

    # Apply SMOTE per fold
    X_tr_res, y_tr_res = smote.fit_resample(X_tr, y_tr)
    lgbm.fit(X_tr_res, y_tr_res)

    y_pred = lgbm.predict(X_val)
    y_proba = lgbm.predict_proba(X_val)[:, 1]

    metrics_cv['Accuracy'].append(accuracy_score(y_val, y_pred))
    metrics_cv['F1-Score'].append(f1_score(y_val, y_pred))
    metrics_cv['ROC-AUC'].append(roc_auc_score(y_val, y_proba))

print("\nCross-Validation Results:")
for metric, scores in metrics_cv.items():
    print(f"{metric}: {np.mean(scores):.4f} ± {np.std(scores):.4f}")
```

```python
# Feature Importance
if best_model and hasattr(best_model, 'feature_importances_'):
    importances = best_model.feature_importances_
    feature_df = pd.DataFrame({
        'Feature': X_train.columns,
        'Importance': importances
    }).sort_values('Importance', ascending=False)

    plt.figure(figsize=(12,8))
    sns.barplot(x='Importance', y='Feature', data=feature_df.head(15))
    plt.title('Top 15 Features')
    plt.tight_layout()
    plt.savefig('feature_importance.png')
    plt.show()

# Save results and model
results_df = pd.DataFrame(results).T
results_df.to_csv('model_results.csv')

import joblib
joblib.dump(best_model, 'best_model.pkl')
print(f"Saved best model ({best_model_name}) with F1: {best_f1:.4f}")
```
